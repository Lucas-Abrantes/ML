#### Questão 11 - Obtenha uma tabela que contenha a variação ano a ano no esta do Rio de Janeiro.

```python
df2['Mes'] = df2['DATA FINAL'].apply(lambda x: x.month)
df_rio = df2[df2['ESTADO'] == 'RIO DE JANEIRO']
```

```python
df_rio
```

```python
df_month_rio = df_rio.groupby('ANO-MES')[['PREÇO MÉDIO REVENDA', 'Mes']].last()
```

```python
((df_month_rio[df_month_rio['Mes'] == 12] / df_month_rio[df_month_rio['Mes'] == 12].shift(1) - 1) * 100).rolling(3).sum()
```

#### Questão 12 - Criar uma tabela contendo uma série temporal com a diferenã absoluta e percentual entre os valores mais baratos e mais caros. Também, fornecer os estados que apresentaram maiores e menores preços.

```python
df_max = df2.groupby('ANO-MES').max()['PREÇO MÉDIO REVENDA']
```

```python
df_min = df2.groupby('ANO-MES').min()['PREÇO MÉDIO REVENDA']
```

```python
df_diff = pd.DataFrame()
df_diff
```

```python
df_diff['abs_diff'] = df_max - df_min
df_diff
```

```python
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
df_diff
```

```python
idx_max = df2.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmax()
idx_min = df2.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
df_diff['MAX'] = df_max
df_diff['MIN'] = df_min
```

```python
df_diff
```

```python
df2.loc[idx_max, :]['ESTADO'].values

# caso os dados não encaixem, ou seja, devido possuir o mesmo tamanho, porém índices diferentes, utilizamos o values para resolver o problema
```

```python
df_diff['ESTADO-MAX'] = df2.loc[idx_max, :]['ESTADO'].values
df_diff['ESTADO-MIN'] = df2.loc[idx_min, :]['ESTADO'].values
```

```python
df_diff
```

```python
df_diff[['ESTADO-MAX']].value_counts()
```

```python
df_diff[['ESTADO-MIN']].value_counts()
```
