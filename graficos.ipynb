```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)

y = x ** 2

plt.plot(x, y, "red")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Função quadrática")
plt.show()
```

```python
plt.subplot(1,2,1)
# numero de linhas, numero de eixos e o gráfico que eu quero trabalhar
plt.plot(x, y, "r--")

plt.subplot(1,2,2)
plt.plot(x, y, "g*-")
```

#### Orientação a objetos

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#esquerda, inferior, largura, altura 
axes.plot(x, y, "r")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Primeiro")




axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

axes2.plot(x, y, "g")
axes2.set_xlabel("Label x")
axes2.set_ylabel("Label y")
axes2.set_title("Segundo")
```

#### Legendas, rótulos e títulos

```python
fig = plt.figure(figsize=(3,3))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, "b--", label="x ** 2")
#ax.plot(x, x**3, "g*-", label="x ** 3")
ax.plot(x, x**3, label="x ** 3", color="#000")


ax.set_title("Várias funções")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")

ax.legend(loc=6)
#loc vai especificar onde eu quero colocar a legenda
```

#### Outro formatos

```python
fig, ax = plt.subplots(figsize=(6,6))


ax.plot(x, x + 1, color="red", linewidth=0.25)
ax.plot(x, x + 2, color="red", linewidth=0.50)
ax.plot(x, x + 3, color="red", linewidth=0.75)
ax.plot(x, x + 4, color="red", linewidth=1.0)



# estilos de linha
# lw é a abreviação de linewidth

ax.plot(x, x + 5, color="green", lw=0.25, linestyle="-")
ax.plot(x, x + 6, color="green", lw=0.25, linestyle="-.")
ax.plot(x, x + 7, color="green", lw=0.25, linestyle=":")


#traços estilizados

line, = ax.plot(x, x + 8, color="black", lw = 1.50) 
line.set_dashes([5, 10, 15, 10]) #Formato: comprimento da linha, comprimento do espaço ...



#simbolos de marcação = '+', 'o', '*', 's', ','  //    '.',  '1', '2', '3'...

ax.plot(x, x + 9, color="blue", lw=3, ls="-", marker= "+")
ax.plot(x, x + 10, color="blue", lw=3, ls="--", marker= "o")
ax.plot(x, x + 11, color="blue", lw=3, ls="-", marker= "s")
ax.plot(x, x + 12, color="blue", lw=3, ls="--", marker= "1")


# tamanho da cor do marcador

ax.plot(x, x + 13, color="purple", lw=1, ls="-", marker= "o", markersize = 2)
ax.plot(x, x + 14, color="purple", lw=1, ls="-", marker= "o", markersize = 4)
ax.plot(x, x + 15, color="purple", lw=1, ls="-", marker= "o", markersize = 8)
ax.plot(x, x + 16, color="purple", lw=1, ls="-", marker= "s", markersize = 8)


```

```python

fig, axes = plt.subplots(1,3, figsize=(7,5))
plt.tight_layout()

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title("Intervalo de eixos")


axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis("tight") # comprimir os valores
axes[1].set_title("Intervalo de eixos")


axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0,60]) # limite do intervalo, de 0 a 60
axes[2].set_xlim([0,60])
axes[2].set_title("Intervalo de eixos")
```
