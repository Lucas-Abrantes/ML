```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'], kde = True, bins = 20)

# kde - exibir a linha do grÃ¡fico
# bins - nÃºmero de elementos do intervalo
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, kind='reg')

# relaciona dois elementos do DataFrame
# passo duas strings como parÃ¢metro
```

```python
sns.pairplot(tips, hue= 'sex', palette='coolwarm')

# grÃ¡fico que relaciona elementos nÃºmericos
# passo um parÃ¢metro categÃ³rico para relacionar com os demais dados
# palette fornece estilo para grÃ¡ficos
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])

#cria uma distribuiÃ§Ã£o normal em torno do eixo e soma
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std)

# gera grÃ¡ficos em barra
# estimator - relaciona as variÃ¡veis ao desvio padrÃ£o
```

```python
sns.set_style('darkgrid')
#sns.set_style('dark')
#sns.set_style('white')
sns.countplot(x='sex', data=tips, palette='deep')

# grÃ¡fico de contagem para uma determinada variÃ¡vel
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, palette='rainbow', hue='sex')
```

```python
sns.boxplot(y='day', x='total_bill', data = tips, palette='rainbow', hue='sex', orient= 'h')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow', split=True, hue='sex')

# diminui pela metado o grÃ¡fico
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

#### Plots de regressÃ£o

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex',  markers = ['s', 'v'], scatter_kws= {'s': 100})

# markers - sÃ£o marcadores para relacionar com os dados
# scatter_kws - passo argumentos a mais apara alterar elementos do grÃ¡fico
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', col = 'time')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', col = 'time', row='day')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex',  markers = ['x', 'v'], scatter_kws= {'x': 100}, aspect = 1.6, size = 1)
```

### Plots matriciais

```python
dados = sns.load_dataset('flights')
dados
```

```python
dados.pivot_table(index='year', columns='month', values='passengers')
```

```python
dados_matrix = dados.pivot_table(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(dados_matrix)

# mapa de calor - utiliza variÃ¡veis categÃ³ricas nas duas dimnesÃµes. JÃ¡ na terceira, utiliza uma cor
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax =  plt.subplots(figsize=(12,8))
sns.heatmap(dados_matrix, ax= ax, annot=True, fmt='.0f', linewidths=1, cmap = 'magma')
```

```python
sns.clustermap(dados_matrix)

# grÃ¡fico que reÃºne elementos do mesmo cluster
```
